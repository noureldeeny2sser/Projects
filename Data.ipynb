```python
import pandas as pd
import numpy as np

# إنشاء داتا "مبهدلة" لمبيعات المتجر
data = {
    'Order_ID': [1001, 1002, 1003, 1004, 1004, 1006, 1007],
    'Product': ['Whey Protein', 'Creatine', 'Dumbbells', 'Vitamin D', 'Vitamin D', 'Whey Protein', 'Creatine'],
    'Price': [1500, 800, np.nan, 400, 400, 1500, 850],
    'Quantity': [2, 1, 1, 3, 3, 1, np.nan],
    'Branch': ['Cairo', 'Alex', 'Alex', 'Cairo', 'Cairo', 'Giza', 'Alex']
}

df_store = pd.DataFrame(data)
print("الجدول قبل أي تنظيف:")
df_store
```

```python
df_store.drop_duplicates(inplace=True)
#Remove duplicates
df_store['Quantity']=df_store['Quantity'].fillna(1)
#fill empty cells in Quantity
df_store['Price']=df_store['Price'].fillna(1000)
#fill empty cells in Price
df_store['Total_Sales']=df_store["Price"]*df_store['Quantity']
#Make a column with Total_Sales
# تجميع المبيعات حسب الفرع ورسمها
df_store.groupby('Branch')['Total_Sales'].sum().plot(kind='bar', color='skyblue', title='Total Sales per Branch')
```
